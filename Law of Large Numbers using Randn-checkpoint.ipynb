```python
import numpy as np

np.set_printoptions(precision=2) #only want 2 decimal place print precision

devs = [] #set empty array to store std deviations after each iteration

N = range(1000,100001,1000)

for n in N: #set number of data points for each iteration
    count = 100/n #create percentile counter
    sigma = [0,0,0,0] #create array to store std deviations
    
    dist = np.random.randn(n) #generate normally distributed data set of length n
    
    for x in dist: #determine whether the data point lies within 1, 2 or 3 standard deviations from the mean
        if abs(x) < 1:
            sigma[0] += count
        elif abs(x) < 2:
            sigma[1] += count
        elif abs(x) < 3:
            sigma[2] += count
        else:
            sigma[3] += count
        
        
    devs.append(sigma) #store data as a new row within the 2D array
```

```python
devs = np.asarray(devs) #convert to numpy array for easier manipulation

for i in range (1,4): #add column values to ones before to get total deviation from mean
        devs[:,i] = (devs[:,i-1] + devs[:,i])
```

```python
import matplotlib.pyplot as plt
plt.close("all")

f, ax = plt.subplots(3, sharex=True) #create 3 subplots that share xaxis
f.subplots_adjust(hspace = 0.7) #adjust space between plots

theory = np.array([68.27,95.45,99.73,99.99]) #expected values of standard deviation for normally distributed data sets

#plot expected value for std dev as horizontal line in each subplot
ax[0].axhline(y=theory[0],color = "y", linestyle = "-")    
ax[1].axhline(y=theory[1],color = "y", linestyle = "-")
ax[2].axhline(y=theory[2],color = "y", linestyle = "-")

#plot std dev against system size for 1, 2, and 3 std devs in separate subplots
ax[0].plot(N,devs[:,0],"c")
ax[1].plot(N,devs[:,1],"b")
ax[2].plot(N,devs[:,2],"g")

ax[0].set_title('1 standard deviation')
ax[1].set_title('2 standard deviations')
ax[2].set_title('3 standard deviations')

#write explanation of what code does and what graphs show
print('\033[1m' + 'Law of Large Numbers' + '\033[0m') #bold title
print("")
print("The law of large numbers states that the mean value of a set will converge to the expected value as the system size goes to infinity.")
print("")
print("Using the randn function in numpy, the above code generates normally distributed data sets of various sizes. It then calculates the percentage of data points in that set that fall within 1, 2 and 3 standard deviations.")

plt.show()

print("The above graphs show the percentage of data points contained within 1, 2 and 3 standard deviations plotted against the size of the data set. The expected value for a normal distribution is also plotted as a solid yellow line for each standard deviation.")
print(" ")
print("The graphs show that as system size increases, the data set moves closer to the expected value for a normally distributed data set. As is stated in the law of large numbers.")
```
